```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataset
np.random.seed(42)
x = np.random.rand(100) * 10  # Random values for variable x
y = 2 * x + 1 + np.random.randn(100) * 2  # Linear relationship with some random noise

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Calculate the correlation coefficient
correlation_coefficient = data['X'].corr(data['Y'])

# Set Seaborn style
sns.set(style="whitegrid")

# Create a scatter plot with Seaborn
sns.scatterplot(x='X', y='Y', data=data)
plt.title(f'Scatter Plot with Correlation: {correlation_coefficient:.2f}')
plt.show()
```
